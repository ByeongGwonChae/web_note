일단 모든 포트로 접속이 불가능한지 알아보도록 하자. 포트가 랜덤으로 바뀌지만 500번 정도 접속해보면 101개쯤은 다 접속해볼 수 있지 않을까.

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
bak = [str(x) for x in range(10000,10101)]
count = 0
jungbok = 0
ineed = ''
```

```python
while len(bak)>0:
    r = requests.get('https://webhacking.kr/challenge/web-16/?server=124.52.70.227')
    soup = bs(r.text,'html.parser')
    t = soup.prettify()
    #print(t.split('\n')[15][53:58])
    
    if t.find('unable to connect to'):
        try:
            bak.remove(t.split('\n')[15][53:58])
        except:
            jungbok = 1
    else:
        ineed=soup.prettify()
        print(t)
        print('\n\n')
    
    count += 1
    if jungbok==1:
        print(count, len(bak), t.split('\n')[15][53:58], '중복')
        jungbok = 0
    else:
        print(count, len(bak), t.split('\n')[15][53:58])
```

10000~100100까지의 101개의 포트 전부 접속해봤지만 전부 접속 불가능하고 나온다. <br>그렇다면 이 방법을 포기하고 다른 방법을 찾아볼 수 밖에 없다.
